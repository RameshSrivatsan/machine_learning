Mean Shift Clustering can be divided into the following steps:
1. Find how much to shift a point with respect to all the other points
2. compute the weights by assuming a multivariate gaussian distribution
3. Shift the point
4. Check the convergence
5. Once all the points have converged or if we have reached maximum number of iterations, group the shifted points into clusters.
6. Return the result

```python
class Grouping:
    def __init__(self):
        self.group_distance = 0.2
     
    def eucledian_distance(self,pointa,pointb):
        'Finds the eucledian distance between pointa and pointb'
        if len(pointa) != len(pointb):
            raise Exception('Points have different dimensions, check the dimensions')
        dim = len(pointa)
        distance = 0
        for i in range(dim):
            distance += (pointa[i] - pointb[i])**2
        return distance
    
    
    def group_points(self,points):
        'Assigning point to a group depending on it"s" distance to the group'
        cluster_ids = []
        groups = []
        index = 0
        for point in points:
            nearest_group_index = self.nearest_group(point,groups)
            if nearest_group_index is None:
                groups.append([point])
                cluster_ids.append(index)
                index += 1
            else:
                cluster_ids.append(nearest_group_index)
                groups[nearest_group_index].append(point)
        return cluster_ids
    
    def nearest_group(self,point,groups):
        'Finds the nearest group to a point'
        nearest_group_index = None
        index = 0
        for group in groups:
            distance = self.distance_of_point_group(point,group)
            if distance <= self.group_distance:
                nearest_group_index = index
            index += 1
        return nearest_group_index
    
    def distance_of_point_group(self,point,group):
        'Returns minimum distance to a group'
        minimum = float("inf")
        for pt in group:
            distance = self.eucledian_distance(point,pt)
            if distance < minimum:
                minimum = distance
        return minimum
```

```python
class MeanShiftAlgorithm():
    def __init__(self):
        self.convergence = 0.01
        
    def eucledian_distance(self,pointa,pointb):
        'Finds the eucledian distance between pointa and pointb'
        if len(pointa) != len(pointb):
            raise Exception('Points have different dimensions, check the dimensions')
        dim = len(pointa)
        distance = 0
        for i in range(dim):
            distance += (pointa[i] - pointb[i])**2
        return distance
    
    def kernel(self,distance,bandwidth):
        'Computes the multivariate gaussian kernel for a given point and returns the kernel'
        dim = len(bandwidth)
    
        covariance = np.multiply(np.power(bandwidth,2),np.eye(dim))
        determinant = np.linalg.det(covariance)

        exponent = -0.5*np.sum(np.multiply(np.dot(distance,np.linalg.inv(covariance)),distance),axis=1)
        val = 1/(np.power(2*math.pi,dim/2)*np.power(np.abs(determinant),0.5))*np.exp(exponent)
        return val
    
    def shift_point(self,point,points,bandwidth):
        'Compute the weights and shift the point'
        #numerator
        weights = self.kernel(point-points,bandwidth)
        tiled_weights = np.tile(weights,[len(point),1])
        
        #denominator
        denominator = np.sum(weights)
        shifted_point = np.multiply(tiled_weights.transpose(),points).sum(axis=0)/denominator
        return shifted_point
    
    def clustering(self,points,bandwidth,iterations):
        'Given a set of points, perform mean shift on all points and compute the number of clusters'
        copy_points = points.copy()
        transformed_points = []
        shifting = [True]*len(points)
        
        for _ in range(iterations):
            for i,point in enumerate(copy_points):
                if not shifting[i]:
                    continue
                new_point = copy_points[i]
                new_point_before_transform = new_point
                new_point = self.shift_point(new_point,points,bandwidth)
                distance = self.eucledian_distance(new_point,new_point_before_transform)
                if distance <= self.convergence:
                    shifting[i] = False
                copy_points[i] = new_point
            transformed_points.append(copy_points.copy())
        
        #Clustering
        clustering_of_shifted_points = Grouping()
        clusters = clustering_of_shifted_points.group_points(copy_points)
        return clusters,transformed_points
```

```python
sample = np.concatenate([sample1,sample2,sample3])
bandwidth = [0.5,0.5]
mean_shift = MeanShiftAlgorithm()
clusters,transformed = mean_shift.clustering(sample,bandwidth,100)
```
